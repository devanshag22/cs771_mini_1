```python
import pandas as pd
import numpy as np
```

```python
# read text sequence dataset
train_seq_df = pd.read_csv("datasets/train/train_text_seq.csv")
train_seq_X = train_seq_df['input_str'].tolist()
train_seq_Y = train_seq_df['label'].tolist()

val_seq_df = pd.read_csv("datasets/valid/valid_text_seq.csv")
val_seq_X = val_seq_df['input_str'].tolist()
val_seq_Y = val_seq_df['label'].tolist()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding

# Convert the strings to lists of integers
X_data = [[int(char) for char in sequence] for sequence in train_seq_X]

# Pad sequences to ensure they are all of length 50
X_data = tf.keras.preprocessing.sequence.pad_sequences(X_data, maxlen=50)


# Convert to numpy arrays
X_data = np.array(X_data)
y_data = np.array(train_seq_Y)

# LSTM model
model = Sequential()

# Embedding layer (if you want to use embedding for the 0-9 digits)
model.add(Embedding(input_dim=10, output_dim=32, input_length=50))

# LSTM layer
model.add(LSTM(64, return_sequences=False))  # return_sequences=False for binary classification

# Fully connected output layer
model.add(Dense(1, activation='sigmoid'))  # Sigmoid for binary classification

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Model summary
model.summary()
```

```python
val_X = [[int(char) for char in sequence] for sequence in val_seq_X]
```

```python
# split the data into 5 forms one consisting of 20% of the data 40% of the data 60% of the data 80% of the data and 100% of the data
from sklearn.model_selection import train_test_split
X_train_20, X_test_20, y_train_20, y_test_20 = train_test_split(X_data, y_data, test_size=0.8, random_state=42)
X_train_40, X_test_40, y_train_40, y_test_40 = train_test_split(X_data, y_data, test_size=0.6, random_state=42)
X_train_60, X_test_60, y_train_60, y_test_60 = train_test_split(X_data, y_data, test_size=0.4, random_state=42)
X_train_80, X_test_80, y_train_80, y_test_80 = train_test_split(X_data, y_data, test_size=0.2, random_state=42)
X_train_100 = X_data
y_train_100 = y_data
```

```python
model.fit(X_data, y_data, validation_data=(val_X,val_seq_Y), epochs=100, batch_size=32)
```
